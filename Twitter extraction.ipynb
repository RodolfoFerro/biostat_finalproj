```python
# We import our libs and credentials:
import tweepy
from credentials import *
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# We setup our credentials:
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
api = tweepy.API(auth)
```

```python
# Get tweets and print them in file
#t_user = "EPN"
#t_user = "Conacyt_MX"
#t_user = "cinvestav"
#t_user = "HarvardBiostats"
t_user = "BiologyData"
#t_user = "FerroRodolfo"
tweets = api.user_timeline(screen_name=t_user, count=200)

print(len(tweets))
```

```python
for tweet in tweets:
    print(tweet.text)
    print("Fecha de creación:", tweet.created_at)
    print("Geolocalización:", tweet.geo)
    print("Longitud del tweet:", len(tweet.text), '\n')
```

```python
data = pd.DataFrame()
data['Tweet'] = np.array([tweet.text for tweet in tweets])
data['Longitud'] = np.array([len(tweet.text) for tweet in tweets])
#data['Fecha'] = np.array([tweet.created_at for tweet in tweets])
#data['Geolocalización'] = np.array([tweet.geo for tweet in tweets])
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize=(10,7.5))
plt.hist(data['Longitud'], normed=False, color='lightseagreen', linewidth=1.5, edgecolor='white')
plt.title("Histograma de longitudes de Tweets")
```

```python
# We export our data to an Excel file:
writer = pd.ExcelWriter(t_user + '.xlsx')
data.to_excel(writer,'Sheet1')
writer.save()
```
